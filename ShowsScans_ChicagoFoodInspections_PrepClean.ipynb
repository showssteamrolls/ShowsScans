**General Cleaning, Converting Data Types:**

```python
main_df.columns=(main_df.columns.str.strip().str.lower().str.replace(r'[^a-z0-9]+','_',regex=True).str.strip('_'))
main_df.head(1)
```

```python
text_cols=['dba_name','aka_name','facility_type','risk','address','city','state','inspection_type','results','violations','location']
main_df[text_cols]=main_df[text_cols].astype('string').apply(lambda s:s.str.strip())

text_cols = ['dba_name','aka_name','facility_type','risk','address','city','state',
             'inspection_type','results','violations','location']
main_df[text_cols] = main_df[text_cols].astype('string').apply(lambda s: s.str.strip())

main_df['inspection_date'] = pd.to_datetime(main_df['inspection_date'], errors='coerce')
main_df['license'] = pd.to_numeric(main_df['license'], errors='coerce').astype('Int64')
main_df['zip'] = main_df['zip'].astype('string').str.extract(r'(\d{5})')[0]
main_df['latitude'] = pd.to_numeric(main_df['latitude'], errors='coerce')
main_df['longitude'] = pd.to_numeric(main_df['longitude'], errors='coerce')
main_df[text_cols + ['zip']] = main_df[text_cols + ['zip']].fillna('_')
```

```python
main_df['year'] = main_df['inspection_date'].dt.year
main_df['month'] = main_df['inspection_date'].dt.month
main_df['month_name'] = main_df['inspection_date'].dt.month_name()
main_df['year_month'] = main_df['inspection_date'].dt.to_period('M').astype(str)
```

**Time Fields:**

```python
main_df['year'] = main_df['inspection_date'].dt.year
main_df['month'] = main_df['inspection_date'].dt.month
main_df['month_name'] = main_df['inspection_date'].dt.month_name()
main_df['year_month'] = main_df['inspection_date'].dt.to_period('M').astype(str)
```

**Normalized Categories:**

```python
results_col = main_df['results'].str.lower()
main_df['results_norm'] = np.select(
    [results_col.eq('pass') | results_col.eq('passed'),
     results_col.str.contains('cond', na=False),
     results_col.str.contains('fail', na=False)],
    ['Pass','Conditional Pass','Fail'],
    'Other'
)
risk_col = main_df['risk'].str.lower()
main_df['risk_norm'] = np.select(
    [risk_col.str.startswith('risk 1') | risk_col.str.contains('high', na=False),
     risk_col.str.startswith('risk 2') | risk_col.str.contains('medium', na=False),
     risk_col.str.startswith('risk 3') | risk_col.str.contains('low', na=False)],
    ['High','Medium','Low'],
    '_'
)
```

**Violations Features:**

```python
violations_col = main_df['violations'].fillna('').astype(str).str.strip()
violations_col = violations_col.str.replace(';','|').str.replace(',','|')
main_df['violations_count'] = np.where(
    violations_col.eq(''), 0, violations_col.str.split('|').apply(lambda L: len([x for x in L if x.strip()]))
)
```

```python
def extract_first_violation_code(violation_entry):
    if not isinstance(violation_entry, str) or not violation_entry.strip():
        return '_'
    for token in violation_entry.split():
        token_lower = token.lower()
        if token_lower.startswith('code') or token[:1].isdigit():
            return token.title()
    return '_'
```

```python
main_df['top_violation_code'] = main_df['violations'].apply(extract_first_violation_code)
```

**Outcomes & Cadence:**

```python
main_df['pass_flag'] = main_df['results_norm'].isin(['Pass','Conditional Pass']).astype('int8')
main_df['fail_flag'] = main_df['results_norm'].eq('Fail').astype('int8')
```

```python
main_df = main_df.sort_values(['license','inspection_date'])
main_df['days_since_prev'] = (main_df.groupby('license', dropna=False)['inspection_date'].diff().dt.days)
```

```python
main_df.to_csv('chicago_food_inspections.csv', index=False)
```
